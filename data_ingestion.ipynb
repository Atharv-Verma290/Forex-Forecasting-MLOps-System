```python
import schedule
import time 

def print_message():
    print("Task executed at:", time.strftime("%H:%M:%S"))

schedule.every(5).seconds.do(print_message)

while True:
    schedule.run_pending()
    time.sleep(1)
```

```python
from supabase import create_client, Client
from dotenv import load_dotenv
import os
load_dotenv()

url = os.environ.get("SUPABASE_URL")
key = os.environ.get("SUPABASE_KEY")
supabase: Client = create_client(url, key)

raw_data = supabase.table("EUR_USD_raw").select("*").order("datetime", desc=True).execute()
print(f"Recieved raw_data from supabase of size: {len(dict(raw_data))}")
```
